```python
import pandas as pd 
import numpy as np
import warnings
```

## Importing dataset

```python
df=pd.read_csv("Downloads/datset - Sheet1.csv")
```

```python
df
```

## DATA CLEANING

```python
print(df.isnull().sum())
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['CustomerName'] = df['CustomerName'].fillna('Unknown')
```

```python
df.loc[df['UnitPrice'].isna() & df['TotalSales'].notna() & df['Quantity'].notna(), 'UnitPrice'] = \
    df['TotalSales'] / df['Quantity']
```

```python
df.loc[df['TotalSales'].isna() & df['UnitPrice'].notna() & df['Quantity'].notna(), 'TotalSales'] = \
    df['UnitPrice'] * df['Quantity']
```

```python
df.loc[df['Quantity'].isna() & df['TotalSales'].notna() & df['UnitPrice'].notna(), 'Quantity'] = \
    df['TotalSales'] / df['UnitPrice']
```

```python
print(df.isnull().sum())
```

```python
df
```

```python
warnings.filterwarnings("ignore")
df['Quantity'].fillna(df['Quantity'].median(), inplace=True)
df['UnitPrice'].fillna(df['UnitPrice'].median(), inplace=True)
df['TotalSales'].fillna(df['TotalSales'].median(), inplace=True)
```

```python
print(df.isnull().sum())
```

## Calcuations

### Sum of sales by region

```python
sales_by_region = df.groupby('Region')['TotalSales'].sum().reset_index()
print(sales_by_region)
```

### Average sales per product

```python
average_sales_by_product = df.groupby('Product')['TotalSales'].mean().reset_index()
print(average_sales_by_product)
```

### Highest & lowest selling products

```python
highest_selling = df.groupby('Product')['Quantity'].sum().reset_index().iloc[0]
print(highest_selling)
```

```python
lowest_selling =df.groupby('Product')['Quantity'].sum().reset_index().iloc[-1]
print(lowest_selling)
```
